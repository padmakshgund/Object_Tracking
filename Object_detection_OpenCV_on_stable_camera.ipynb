```python
import cv2
import matplotlib.pyplot as plt
from tracker import *
```

```python
## Background Subtractor method to track the moving object by cetroid method
tracking = EuclideanDistTracker()
video = cv2.VideoCapture("object_tracking/highway.mp4")
object_detector = cv2.createBackgroundSubtractorMOG2(history = 100,varThreshold = 40)
while video.isOpened():
    ret,frame = video.read()
#     break
    if ret:
        # Region of interest
        roi = frame[320:610,550: 750]
        mask = object_detector.apply(roi)
        _,thresh = cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
        countour,_ = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
        detector = []
        for contour in countour:
            (x,y,w,h) = cv2.boundingRect(contour)
            contour_area = cv2.contourArea(contour)
            if contour_area > 500:
                cv2.drawContours(thresh,contour,-1,(0,255,0))
                cv2.rectangle(roi,(x,y),(x+w,y+h),(0,255,0),4)
                detector.append([x,y,w,h])
#                 print(detector)
        box_id = tracking.update(detector)
        for i in box_id:
            cv2.putText(roi,str(i[-1]),(i[0]-10,i[1]-10),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
        cv2.imshow("frame",frame)
        cv2.imshow("roi",roi)
        if cv2.waitKey(1) & 0xFF ==ord("q"):
            break
    else:
        break
        
video.release()
cv2.destroyAllWindows()
```
